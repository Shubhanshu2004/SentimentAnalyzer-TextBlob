```python
# ğŸ“¦ Import the required library
from textblob import TextBlob

# ğŸ§  Define a function for sentiment analysis
def analyze_sentiment():
    # âœï¸ User se sentence input lena
    text = input("Enter your sentence: ")

    # ğŸ“ Create a TextBlob object
    blob = TextBlob(text)

    # ğŸ“Š Extract polarity and subjectivity
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    # ğŸ“¢ Print the result
    print("\n--- Sentiment Analysis Result ---")
    print("Polarity:", polarity)
    print("Subjectivity:", subjectivity)

    # ğŸ˜€ ğŸ˜ ğŸ˜ Emoji-based sentiment output
    if polarity > 0:
        print("Sentiment: Positive ğŸ˜€")
    elif polarity < 0:
        print("Sentiment: Negative ğŸ˜")
    else:
        print("Sentiment: Neutral ğŸ˜")

# â–¶ï¸ Call the function
analyze_sentiment()
```
